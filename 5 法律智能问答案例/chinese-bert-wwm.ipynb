## 训练

```python
!python baseline/run_cail.py \
    --name chinese-bert-wwm \
    --bert_model ./models/chinese_wwm_pytorch \
    --data_dir ./output/data/chinese-bert-wwm \
    --batch_size 2 \
    --eval_batch_size 32 \
    --lr 1e-5 \
    --gradient_accumulation_steps 4 \
    --seed 56 \
    --epochs 25
```
